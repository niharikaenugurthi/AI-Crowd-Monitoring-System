```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#ways to create arrays
a = np.array([1,2,3])
b=np.zeros(3) #[0 0 0]
c=np.ones(3) #[1 1 1]
d =np.arange(1,10) #[1 2 3 ... 9]
e =np.linspace(0,1,5) #5 values between 0 and 1
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
#indexing
a= np.array([10,20,30,40])
print(a[0]) #10
print(a[-1]) #40
```

```python
#slicing
a = np.array([1,2,3,4,5])
print(a[1:4]) #[2 3 4]
```

```python
# operations (no loops needed) important
a= np.array([1,2,3])
print(a+2) #[3 4 5]
print(a*2) #[2 4 6]
```

```python
#shape (important)
a = np.array([[1,2,3],[4,5,6]])
print(a.shape) # (2,3)
```

```python
#reshape 1D to 2D (very imp)
arr = np.arange(1,7)
new_arr = arr.reshape(2,3) #2 rows, 3 columns

print(new_arr)
```

```python
#Flattern (Back to 1D)
arr = np.arange(1,7)
new_arr = arr.reshape(2,3)
print(new_arr.flatten())
```

```python
#2D Indexing
arr = np.array([[10,20,30],[40,50,60]])
print(arr[0,1]) #20
print(arr[1,2]) #60

#Row & Column Access
print(arr[0]) #first row
print(arr[:,1]) #second column
print(arr[:,2])
```

```python
#important functions
arr = np.array([1,2,3,4,5])
print(np.sum(arr)) #15
print(np.mean(arr)) #3.0
print(np.max(arr)) #5
print(np.min(arr)) #1
```

```python
#Broadcasting (Power Topic)
arr = np.array([1,2,3])
print(arr + 5)
```

```python
#Boolean Filtering
arr = np.array([10,20,30,40])
print(arr>20) #[False False True True]
print(arr[arr>20]) #[30 40]
```

```python
#pandas = data handling tool(tables like excel)
#Example table:

#name	age
# A	  20
# B	  25
```

```python
#create DataFrame
import pandas as pd
df = pd.DataFrame({
    "name":["A","B","C"],
    "age":[20,25,30]
})
print(df)

#View Data
print(df.head()) #First rows
print(df.tail()) #Last rows
print(df.shape) #(rows,colums)
print(df.columns) #column names
```

```python
#select columns
print(df["name"]) #single column
print(df[["name","age"]]) #multiple colums
```

```python
#filtering
print(df[df["age"]>20])
```

```python
#Add New Column
df["double_age"] = df["age"]*2
print(df)
```

```python
#Basic Info
df.info()
df.describe()
```

```python
#Sorting
df.sort_values("salary")
df.sort_values("salary",ascending=False)
```

```python
#groupby() very impotantt
#used to group data and apply operations
import pandas as pd
df = pd.DataFrame({
    "name":["A","B","C","D"],
    "dept":["IT","HR","IT","HR"],
    "salary":[5000,4000,6000,4500]
})

print(df.groupby("dept")["salary"].mean())


#multiple conditions
df[(df["salary"] > 4000) & (df["dept"] == "IT")]


#real data style
df["bonus"] = df["salary"]*0.1

```

```python
#Missing values
df = pd.DataFrame({
    "name": ["A","B","C"],
    "age": [20,None,30]
})

#check missing
df.isnull()
```

```python
#remove missing
df.dropna()
```

```python
#fill missing
df.fillna(0)
```
